```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------- SAMPLE DATASET ----------------
data = {
    "EmployeeID": [101, 102, 103, 104, 105, 106, 107, 108],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Finance", "IT", "HR"],
    "Salary": [50000, 60000, 55000, 62000, 58000, 57000, 61000, 59000]
}
df = pd.DataFrame(data)
num_col = "Salary"

# ---------------- NUMERICAL ANALYSIS ----------------
print(f"\nStatistics for {num_col}:")
print("Mean:", df[num_col].mean())
print("Median:", df[num_col].median())
print("Mode:", df[num_col].mode()[0])
print("Standard Deviation:", df[num_col].std())
print("Variance:", df[num_col].var())
print("Range:", df[num_col].max() - df[num_col].min())

# Histogram
plt.figure(figsize=(8,4))
sns.histplot(df[num_col], bins=5, kde=True, color='skyblue')
plt.title(f"Histogram of {num_col}")
plt.xlabel(num_col)
plt.ylabel("Frequency")
plt.show()

# Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x=df[num_col], color='lightgreen')
plt.title(f"Boxplot of {num_col}")
plt.show()

# ---------------- CATEGORICAL ANALYSIS ----------------
cat_col = "Department"
counts = df[cat_col].value_counts()
print(f"\nCategory Counts for {cat_col}:")
print(counts)

# Bar chart
plt.figure(figsize=(8,4))
sns.barplot(x=counts.index, y=counts.values, palette='pastel')
plt.title(f"Bar Chart of {cat_col}")
plt.xlabel(cat_col)
plt.ylabel("Count")
plt.show()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct="%1.1f%%", colors=sns.color_palette('pastel'))
plt.title(f"Pie Chart of {cat_col}")
plt.show()
```
